# Supervised Learning: Regression Models

## Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

**Answer:**

Simple Linear Regression (SLR) is a supervised learning technique used to understand the relationship between one independent variable (X) and one dependent variable (Y). It assumes that the relationship between X and Y can be represented using a straight line. The main purpose of SLR is to predict the value of the dependent variable based on a given value of the independent variable and to measure how changes in X affect Y.

## Question 2: Key assumptions of Simple Linear Regression

**Answer:**

1. Linearity – The relationship between X and Y must be linear.
2. Independence – Observations should be independent of each other.
3. Homoscedasticity – The variance of residuals should remain constant.
4. Normality – Residuals should be normally distributed.
5. No extreme outliers – Outliers should not heavily influence the model.

## Question 3: Mathematical equation of Simple Linear Regression

**Answer:**

The mathematical equation of a simple linear regression model is:

Y = β₀ + β₁X + ε

Where:
- Y is the dependent variable
- X is the independent variable
- β₀ is the intercept
- β₁ is the slope coefficient
- ε is the error term

## Question 4: Real‑world example of Simple Linear Regression

**Answer:**

A real‑world example of simple linear regression is predicting a student’s exam score based on the number of hours studied. Here, study hours act as the independent variable, while exam score is the dependent variable.

## Question 5: Method of Least Squares

**Answer:**

The method of least squares estimates regression coefficients by minimizing the sum of squared differences between actual values and predicted values. This ensures the best possible fit for the regression line.

## Question 6: Logistic Regression vs Linear Regression

**Answer:**

Logistic Regression is used for classification problems with categorical outcomes, while Linear Regression predicts continuous values. Logistic regression uses a sigmoid function, whereas linear regression uses a straight‑line equation.

## Question 7: Regression evaluation metrics

**Answer:**

1. Mean Absolute Error (MAE)
2. Mean Squared Error (MSE)
3. Root Mean Squared Error (RMSE)

## Question 8: Purpose of R‑squared

**Answer:**

R‑squared indicates how much variation in the dependent variable is explained by the independent variable. Higher values mean better model performance.

```python

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

## Question 10: Interpretation of coefficients

**Answer:**

The intercept shows the expected value of Y when X is zero. The slope shows how much Y changes for a one‑unit increase in X, helping us understand the relationship direction and strength.
